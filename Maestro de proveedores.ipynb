```python
db2['Fecha creación solicitud'] = db2['Fecha creación solicitud'].astype('datetime64[ns]', dayfirst=True, format='%d-%m-%Y')
db2['Fecha Cierre Reviewer'] = db2['Fecha Cierre Reviewer'].astype('datetime64[ns]', dayfirst=True, format='%d-%m-%Y')
db2['Fecha creación solicitud'] = db2['Fecha creación solicitud'].astype('datetime64[ns]', dayfirst=True, format='%d-%m-%Y')
```

```python
db2['Month creation request'] = db2['Fecha creación solicitud'].dt.month
```

```python
#https://www.geeksforgeeks.org/working-with-missing-data-in-pandas/
db2['Email implant'].fillna('No email implant', inplace = True)
db2['Fecha Cierre Reviewer'].fillna('Aun en reviewer', inplace = True)
db2['Fecha Presentacion Reviewer'].fillna('Aun en reviewer', inplace = True)
#db2.replace(to_replace = np.nan, value = 0,inplace=True) 
```

```python
Cantidad_de_Lineas = []
dictionary = {}
count=1
```

```python
for i in range(len(db2['BU-PO'])):
    if db2['BU-PO'][i] in dictionary:
        count=dictionary.get(db2['BU-PO'][i])+1
        #Change Values dictionary
        dictionary[db2['BU-PO'][i]]=count
        #Accessing Items dictionary
        x = dictionary.get(db2['BU-PO'][i])
        #db2['Cantidad_de_Lineas'].append(x)
        Cantidad_de_Lineas.append(x)
    else:
        count=1
        #Change Values dictionary
        dictionary[db2['BU-PO'][i]]=count
        #Accessing Items dictionary
        x = dictionary.get(db2['BU-PO'][i])
        #db2['Cantidad_de_Lineas'].append(x)
        Cantidad_de_Lineas.append(x)
```
